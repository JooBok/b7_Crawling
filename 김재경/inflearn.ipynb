# 정적 프로그램

```python
import pandas as pd 
import numpy as np 
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
import re
import time
import requests

inf_res = pd.DataFrame(columns=['강의명', '강사', '가격', '리뷰수'])

# page = 1 ~ 14

for page in range(1, 2):
    in_url = f'''https://www.inflearn.com/courses?s=aws&order=search&page={page}'''
    driver = webdriver.Chrome()
    driver.get(in_url)

    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    price = []
    review = []

    title = [a.text for a in soup.select('div.course_title')]
    inst = [b.text for b in soup.select('div.instructor')]

    for review_num in range(1, 25):
        try:
            review.append(int(re.sub('\D', '', driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{review_num}]/div/a/div[2]/div[3]/span').text)))
        except:
            review.append(None)

    for c in soup.select('div.price'):
        try:
            price.append(int(re.sub('\D', '', c.text)))
        except:
            price.append(0)

    inf_temp = pd.DataFrame([title, inst, price, review], index = ['강의제목', '강사', '가격', '리뷰수']).T
    inf_res = pd.concat([inf_res, inf_temp])

inf_res
```

```python
inf_res2 = pd.DataFrame(columns= ['강의명', '수강생', '기술', '별점', '목차', '강의내용'])

for page in range(1, 2):
    in_url = f'''https://www.inflearn.com/courses?s=aws&order=search&page={page}'''
    driver = webdriver.Chrome()

    lec_index = []

    for lect in range(1,3):
        try:
            driver.get(in_url)
            driver.get(in_url)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            time.sleep(0.5)
            driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{lect}]').click()
            time.sleep(0.5)
            driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{lect}]').click()
            time.sleep(0.5)
            driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{lect}]').click()
    
            time.sleep(1)
            # 강의 제목 (merge 용)
            title = driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[2]/h1').text.split('\n')[0]
            time.sleep(0.1)
            # 별점
            star = re.sub('\s', ' ', driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text).split(' ')[0]
            time.sleep(0.1)
            # 수강생
            stu = re.sub('\s', ' ', driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text).split(' ')[-2]+re.sub('\s', ' ', driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text).split(' ')[-1]
            time.sleep(0.1)
            # 기술
            tec = re.sub('\s', ', ', driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[5]').text)
            time.sleep(0.1)
            # 강의 내용
            cont = driver.find_element(By.XPATH, '//*[@id="description"]/section[1]/div[2]').text.replace('\n', ', ')


            # 목차
            # index_nums = 1 ~ 20 try except pass 사용하여 최대한 가져오기
            for index_nums in range(1, 21):
                try:
                    lec_index.append(driver.find_element(By.XPATH, f'//*[@id="curriculum"]/div[3]/div/div[{index_nums}]/div[1]').text)

                except:
                    pass
            temp_res2 = pd.DataFrame([title, stu, tec, star, lec_index, cont], index= ['강의명', '수강생', '기술', '별점', '목차', '강의내용']).T
            inf_res2 = pd.concat([inf_res2, temp_res2])

        except:
            pass
        time.sleep(1)
    driver.close()
    
    
```

* 클릭은 성공 첫번째 강의 data는 clawling 성공했지만
* 창을 다시 생성후 2번째 강의로 넘어가지 않음

```python
inf_res2 = pd.DataFrame(columns=['강의명', '수강생', '기술', '별점', '목차', '강의내용'])

for page in range(1, 2):
    in_url = f'https://www.inflearn.com/courses?s=aws&order=search&page={page}'
    driver = webdriver.Chrome()

    for lect in range(1, 3):
        try:
            driver.get(in_url)
            time.sleep(2)  
            driver.find_element(By.XPATH, f'//*[@id="courses_section"]/div/div/div/main/div[4]/div/div[{lect}]').click()
            time.sleep(2)  

            # 강의 정보 수집
            title = driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[2]/h1').text.split('\n')[0]
            star = int(re.sub('\D', '', driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text.split('\n')[0]))/10
            stu = driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text.split(' ')[-2] + driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[3]').text.split(' ')[-1]
            tec = driver.find_element(By.XPATH, '//*[@id="main"]/section/div[1]/div[1]/div/div/div[2]/div[5]').text.replace('\n', ', ')
            cont = driver.find_element(By.XPATH, '//*[@id="description"]/section[1]/div[2]').text.replace('\n', ', ')

            # 목차 수집
            lec_index = []
            for index_nums in range(1, 21):
                try:
                    lec_index.append(driver.find_element(By.XPATH, f'//*[@id="curriculum"]/div[3]/div/div[{index_nums}]/div[1]').text)
                except:
                    pass

            
            temp_res2 = pd.DataFrame([[title, stu, tec, star, lec_index, cont]], columns=['강의명', '수강생', '기술', '별점', '목차', '강의내용'])
            inf_res2 = pd.concat([inf_res2, temp_res2], ignore_index=True)

        except Exception as e:
            print(f"에러 발생: {e}")
        
        finally:
            driver.quit()
            time.sleep(1)
```

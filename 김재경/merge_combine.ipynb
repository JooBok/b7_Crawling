```python
import pandas as pd
import numpy as np
import re
```

```python
data = pd.read_csv('wanted_data.csv', index_col=0)
mer1 = pd.read_excel('corp_data_1.xlsx', index_col=0)
mer2 = pd.read_csv('corp_data_2.csv', index_col=0)
```

```python
data.기업명 = data.기업명.apply(lambda x: re.sub(r'\([^)]*\)','',x))
data.기업명 = data.기업명.apply(lambda x: re.sub('\\W', '', x))
mer1.기업명 = mer1.기업명.apply(lambda x: re.sub(r'\([^)]*\)','',x))
mer1.기업명 = mer1.기업명.apply(lambda x: re.sub('\\W', '', x))
mer2.기업명 = mer2.기업명.apply(lambda x: re.sub(r'\([^)]*\)','',x))
mer2.기업명 = mer2.기업명.apply(lambda x: re.sub('\\W', '', x))
```

# mer 2 전처리

```python
mer2
```

```python
# 설립연도 전처리

mer2.설립연도 = mer2.설립연도.apply(lambda x: ''.join(re.findall('\d\d\d\d', x)))
```

```python
# 매출액 전처리

sales_temp = []

for sales in mer2.매출액:
    if sales != '-':
        sales_temp.append(re.sub(r'\s\(\d\d\d\d\)', '', sales.replace(',','')))
    else:
        sales_temp.append(None)
sales_data = []

for sales in sales_temp:
    if sales != None:
        sales_data.append(re.sub('\s', '', sales))
    else:
        sales_data.append(None)

sales_data_res = []
try:
    for q in sales_data:
        if q != None:    
            if '조' in q:
                    sales_data_res.append(int(''.join(re.findall(r'(\d+)조', q)))*100000000 + int(''.join(re.findall(r'(\d+)억', q)))*10000 + int(''.join(re.findall(r'억(\d+)', q))))
            elif '억원' in q:
                 sales_data_res.append(int(''.join(re.findall(r'(\d+)억원', q)))*10000)
            elif '억' in q:
               sales_data_res.append(int(''.join(re.findall(r'(\d+)억', q)))*10000 + int(''.join(re.findall(r'억(\d+)', q))))   
            else:
                sales_data_res.append(int(''.join(re.findall(r'(\d+)만원', q))))
        else:
             sales_data_res.append(None)
except:
     print(q)
mer2.매출액 = sales_data_res
```

```python
# 사원수 전처리

mer2.사원수 = mer2.사원수.apply(lambda x: re.sub('\(헤드헌터 1 명\)', '', x))

emp_temp = []

for emp in mer2.사원수:
    if emp != '-':
        emp_temp.append(int(''.join(re.findall('\d', emp))))
    else:
        emp_temp.append(None)
mer2.사원수 = emp_temp
```

```python
# 업종분류를 표준산업분류로 크롤링하지 않아 드랍 컬럼

mer2.drop(columns = ['업종분류'], inplace=True)
```

```python
# 주소 전처리

mer2.주소.fillna(2, inplace = True)

address_temp = []
for addr in mer2.주소:
    if addr != 2:
        address_temp.append(' '.join(addr.split()[:2]))

    else:
        address_temp.append(None)
mer2.주소 = address_temp
```

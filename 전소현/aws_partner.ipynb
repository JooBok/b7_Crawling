```python
import requests
from bs4 import BeautifulSoup

all_titles=[] #회사이름
all_absts=[] #요약
    
for page in range(1, 91):
    url = f'https://partners.amazonaws.com/ko/search/partners?page={page}'
  
    
    response = requests.get(url)
    soup = BeautifulSoup(driver.page_source,'html.parser')
    titles = [t.text for t in soup.select('span.psf-partner-search-details-card__title')]
    all_titles.extend(titles)

    absts = [a.text for a in soup.select('p.psf-partner-search-details-card__description')]
    all_absts.extend(absts)

# 회사 이름과 요약을 함께 출력
for title, abst in zip(all_titles, all_absts):
    print("Company:", title)
    if abst:  # 요약 정보가 있는 경우에만 출력
        print("Abstract:", abst)
    else:
        print("Abstract: No information available")  # 요약 정보가 없는 경우에는 해당 메시지 출력
```

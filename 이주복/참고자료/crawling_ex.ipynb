```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd
import re
```

```python
book_res = pd.DataFrame(columns=['제목','연도','저자','대출현황','요약'])
```

```python
page_no = 1
page_url = f'http://kyobobook.sogang.ac.kr/elibrary-front/search/searchList.ink?brcd=&sntnAuthCode=&contentAll=&cttsDvsnCode=&orderByKey=&schClst=all&schDvsn=000&reSch=&ctgrId=&allClstCheck=on&clstCheck=ctts&clstCheck=autr&clstCheck=pbcm&clstCheck=intc_cntt&allDvsnCheck=000&dvsnCheck=001&schTxt=it&reSchTxt=&selViewCnt=20&pageIndex={page_no}&recordCount=20'
```

# 무신사 크롤링

```python
musinsa_result = pd.DataFrame(columns=['제조사','상품명','할인전가격','판매가','만족도'])

max_page = int(soup.select('span.totalPagingNum')[0].text)

if max_page > 400:
    max_page = 400

for page in range(1, max_page+1):
    url = f'''https://www.musinsa.com/categories/item/001005?d_cat_cd=001005&brand=&list_kind=small&sort=pop_category&sub_sort=&page={page}&display_cnt=90&group_sale=&exclusive_yn=&sale_goods=&timesale_yn=&ex_soldout=&plusDeliveryYn=&kids=&color=&price1=&price2=&shoeSizeOption=&tags=&campaign_id=&includeKeywords=&measure='''

    req = requests.get(url)
    content = req.content
    soup = BeautifulSoup(content, 'html.parser')

    brand = [b.text for b in soup.select('p.item_title')]

    name = []

    for n in soup.select('p.list_info'):
        n2 = n.text
        name.append(re.sub('\s+',' ',n2)[1:-1])

    price1 = []
    price2 = []

    for p in soup.select('p.price'):
        p_temp = p.text.split('\n')
        price1.append(p_temp[1])
        price2.append(p_temp[2])

    review = []

    for a in soup.select('div.article_info'):
        # 리뷰가 없는 경우, span.bar 객체가 없습니다
        if a.select('span.bar') == []:
            review.append(float('nan'))
        else:
            # 무신사 리뷰 평점은 span.bar의 길이에 비례
            a2 = a.select('span.bar')[0]['style'][6:]
            review.append(a2)

    musinsa_temp = pd.DataFrame([brand, name, price1, price2, review], 
                                index=['제조사','상품명','할인전가격','판매가','만족도']).T

    def price_change(x):
        if x != x:
            return x
        x2 = re.sub('[ ,원%]+','',x)
        if x2 == '':
            return float('nan')
        else:
            return int(x2)

    musinsa_temp['할인전가격'] = musinsa_temp['할인전가격'].apply(lambda x: price_change(x))
    musinsa_temp['판매가'] = musinsa_temp['판매가'].apply(lambda x: price_change(x))
    musinsa_temp['만족도'] = musinsa_temp['만족도'].apply(lambda x: price_change(x))

    musinsa_result = pd.concat([musinsa_result, musinsa_temp])
```

```python
url = 'https://www.hrd.go.kr/hrdp/ti/ptiao/PTIAO0300L.do?crseTracseSeSelected=%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0&ncsSelected=&areaSelected=%EC%A0%84%EA%B5%AD+%3E+%EC%A0%84%EC%B2%B4&searchTotalKeyword=undefined&crseTracseSe=001%7C%EB%B9%85%EB%8D%B0%EC%9D%B4%ED%84%B0&area=00%7C%EC%A0%84%EA%B5%AD+%EC%A0%84%EC%B2%B4&pageId=6&'
```

```python
req = requests.get(url)
content = req.content
soup = BeautifulSoup(content, 'html.parser')
```

```python
zone = [re.sub('\s+',' ',z.text) for z in soup.select('p.zone')]
address = [re.sub('\s+',' ',s.text)[1:-1].split(' ( ')[0] for s in soup.select('p.school')]
tel = [re.sub('\s+',' ',s.text)[1:-1].split(' ( ')[1][:-2] for s in soup.select('p.school')]
percentage = [re.sub('\s+','',p.text) for p in soup.select('div.bar')]

s_date = []
e_date = []

for d in soup.select('dl.item.term'):
    temp = re.findall('\d+-\d+-\d+',d.text)
    
    if temp == []:
        continue
        
    else:
        s_date.append(temp[0])
        e_date.append(temp[1])
```

```python
pd.DataFrame([zone, address, tel, percentage, s_date, e_date], 
             index=['업체명','지역','전화번호','취업률','시작일','종료일']).T
```
